Ketika menghadapi dataset yang memiliki variabel non-linear, model linear regresi berganda dapat diubah menjadi model regresi non-linear dengan menggunakan teknik transformasi fitur atau polinomial feature.

Polinomial feature adalah teknik yang memperluas fitur-fitur asli dengan menambahkan pangkat dan interaksi dari fitur-fitur tersebut. Dengan menambahkan fitur-fitur polinomial, kita dapat menangkap hubungan non-linear antara variabel input dan variabel target.

Berikut adalah langkah-langkah umum untuk menangani dataset dengan variabel non-linear menggunakan model linear regresi berganda:

1. Mempersiapkan dataset: Persiapkan dataset dengan memisahkan variabel input (X) dan variabel target (Y).

2. Polinomial Feature: Gunakan teknik polinomial feature untuk membuat fitur-fitur polinomial dari variabel input. Ini dapat dilakukan dengan menggunakan library seperti sklearn.preprocessing.PolynomialFeatures. Tentukan derajat polinomial yang sesuai dengan dataset Anda.

3. Normalisasi Fitur: Jika perlu, lakukan normalisasi pada fitur-fitur untuk menghindari masalah skala yang tidak seimbang dalam model. Anda dapat menggunakan metode normalisasi seperti Z-score atau Min-Max Scaling.

4. Membagi Data: Bagi dataset menjadi data latih (training data) dan data uji (testing data) menggunakan metode seperti train-test split atau cross-validation.

5. Pelatihan Model: Latih model regresi linear berganda pada fitur-fitur polinomial yang telah dibuat menggunakan dataset latih. Anda dapat menggunakan library seperti sklearn.linear_model.LinearRegression untuk melatih model.

6. Evaluasi Model: Evaluasi performa model pada dataset uji menggunakan metrik evaluasi seperti mean squared error (MSE), r-squared (R2), atau mean absolute error (MAE).

7. Interpretasi Hasil: Analisis hasil model untuk memahami hubungan antara variabel input dan variabel target. Visualisasikan hubungan tersebut jika diperlukan.

8. Prediksi dengan Data Baru: Setelah melatih model, Anda dapat menggunakannya untuk membuat prediksi pada data baru dengan menerapkan transformasi fitur polinomial yang sama pada data baru.

Sumber referensi:

* Hastie, T., Tibshirani, R., & Friedman, J. (2009). The Elements of Statistical Learning: Data Mining, Inference, and Prediction. Springer Science & Business Media.

* Müller, A. C., & Guido, S. (2016). Introduction to Machine Learning with Python: A Guide for Data Scientists. O'Reilly Media.

```python
# Pilih X dan Y variabel
X = data.drop(['ID','surface_roughness'], axis=1)
Y = data['surface_roughness']
```

## Splitting Data Menjadi Data Train dan Test

```python
# Membagi data menjadi data latihan dan data uji
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print('Data Train : ', end='')
print(X_train.shape, Y_train.shape)
print('Data Testing : ', end='')
print(X_test.shape, Y_test.shape)
```

```python
# Membuat objek PolynomialFeatures dengan derajat polinomial 2
poly_features = PolynomialFeatures(degree=2)
# Melakukan transformasi fitur polinomial pada data latihan
X_poly = poly_features.fit_transform(X)
```

## Modeling

Membuat model Regresi Linier (Linier Regression) kemudian melatihnya menggunakan data train.

```python
model = LinearRegression()
model.fit(X_poly,Y)
```

## 2. Uji normalitas
Disini akan menguji normalitas residual (errornya). Mengapa? karena menurut sifat distribusi normal, kalau errornya normal dan modelnya linier maka koefisien regresinya juga berdistribusi normal (terhindar dari bias outlier).

```python
y_train_predict = model.predict(poly_features.fit_transform(X_train)) 
residuals = Y_train - y_train_predict
normality_train = sm.stats.stattools.jarque_bera(residuals)
print("Jarque-Bera train for normality:")
print("JB statistic:", normality_train[0])
print("p-value:", normality_train[1])

sns.distplot(residuals)
plt.show()
```

Ho : data = berdistribusi normal

Ha : data tidak berdistribusi normal

Karena pvalue > 0.05, maka Ho diterima.

Plot histogram yang membentuk lonceng dan puncaknya cenderung ke tengah juga mendukung bahwa **errornya sudah berdistribusi normal**.

## 3. Tidak terjadi Heteroskedastisitas

Homoskedastisitas  adalah  kondisi ketika  nilai  residu/error  pada  tiap  nilai  prediksi  bervariasi  dan  variasinya  cenderung konstan. Lawan dari homoskedastisitas adalah heteroskedastisitas,Heteroskedastisitas merupakan kondisi dimana error cenderung berubah dan malah berkorelasi dengan prediksinya. Jika ini terjadi tentu menunjukkan bahwa prediksi ini memiliki suatu bias tertentu.

```python
plt.figure(figsize=(8,4))
plt.scatter(y_train_predict, residuals)
plt.show()
```

Terlihat plot error-nya berada di sekitar angka yang sama meski nilai prediksinya bertambah.

Artinya nilai prediksi kita tidak terganggu oleh errornya atau **tidak terjadi heteroskedastisitas**.

```python
lm, lm_p_value, fvalue, f_p_value = het_breuschpagan(residuals, poly_features.fit_transform(X_train))
print('Breusch-Pagan test')
print(f'LM test statistic: {lm:.3f}, p-value: {lm_p_value:.3f}')
print(f'F test statistic: {fvalue:.3f}, p-value: {f_p_value:.3f}')
```

## 4. Tidak terjadi Multikolinearitas (khusus regresi berganda)
Multokolinearitas maksudnya adalah hubungan yang kuat antar feature.

Regresi linier mengasumsikan bahwa feature-featurenya tidak saling berhubungan. Tentu saja ini hanya berlaku untuk regresi linier berganda (yang featurenya lebih dari 1).

Salah 1 cara mengujinya adalah dengan melihat nilai VIF.

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["predictor"] = X.columns

print(vif)
```

Some papers argue that a VIF<10 is acceptable, but others says that the limit value is 5.
- "10" as the maximum level of VIF (Hair et al., 1995)
- "5" as the maximum level of VIF (Ringle et al., 2015)

Berdasarkan kriteria Hair, **tidak terjadi kolinearitas**.

> **Catatan:** Namun, berdasarkan kriteria Ringle masih terjadi kolinearitas. Disini kita menggunakan teori Hair.

# Mengevaluasi hasil regresi linier

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# Predicting on test data-set
y_test_predict = model.predict(poly_features.fit_transform(X_test)) 
# Evaluating the model on test dataset
mse = mean_squared_error(Y_test, y_test_predict)
rmse_test = np.sqrt(mean_squared_error(Y_test, y_test_predict))
r2_test = r2_score(Y_test, y_test_predict)
mae= mean_absolute_error(Y_test, y_test_predict)
mape= mean_absolute_percentage_error(Y_test, y_test_predict)
print("R2 score of testing set is {}".format(r2_test))
print("MSE of testing set is {}".format(rmse_test))
print("RMSE of testing set is {}".format(rmse_test))
print("MAE of testing set is {}".format(mae))
print("MAPE of testing set is {}".format(mape))
```

```python
# Menampilkan nilai aktual dan nilai prediksi dalam bentuk dataframe
df_results = pd.DataFrame({'Actual': Y_test, 'Predicted': y_test_predict})
print("Actual vs Predicted Surface Roughness:")
print(df_results)
```

```python
# Menampilkan grafik nilai aktual dengan nilai prediksi
plt.scatter(df_results.index, df_results['Actual'], color = '#9A32CD', label="Actual")
plt.scatter(df_results.index, df_results['Predicted'], color = '#00EE00', label="Predicted", marker = 'x')
plt.xlabel("Data Points")
plt.ylabel("Surface Roughness")
plt.title("Actual vs Predicted Surface Roughness")
plt.legend()
plt.show()
```

```python
df_results['Percentge'] = df_results.apply(lambda row: row['Actual'] / row['Predicted'] if row['Predicted'] > row['Actual'] else
                                          row['Predicted'] / row['Actual'], axis=1)
df_results
```
